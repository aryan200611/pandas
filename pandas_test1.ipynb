```python
import pandas as pd

# Load dataset
df = pd.read_csv("data.csv")

# Data Cleaning

# Remove '/5' from the rate column
df['rate'] = df['rate'].str.replace('/5', '', regex=False)

# Convert rate to float
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Convert votes to numeric (if needed)
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')

# Remove rows with missing values in rate or votes
df = df.dropna(subset=['rate', 'votes'])

# Filter the data
filtered_df = df[(df['votes'] > 700) & (df['rate'] > 3.8)]

# Print only the required columns
print(filtered_df[['name', 'rate', 'votes']])
```
